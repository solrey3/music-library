```python
url = "https://www.billboard.com/charts/hot-100"
```

```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')

song_table = soup.find('ol', class_='chart-list__elements')

song_table
```

```python
songs = song_table.find_all('li')

songs[0]
```
