```python
my_tags = ['title', 'artist', 'date', 'originaldate', 'album', 'tracknumber', 'discnumber', 'albumartist', 'compilation', 'genre', 'comment', 'length', 'bitrate']
my_tags
```

```python
from mutagen.id3 import ID3
from mutagen.mp3 import MP3
from mutagen.easyid3 import EasyID3
from pathlib import Path
import sys

df = pd.DataFrame(columns=my_tags)

path_of_the_directory= 'test_audio'
# path_of_the_directory= '/mnt/e/Jukebox/Music'
# path_of_the_directory= '/mnt/e/Jukebox/Music/Cutty Ranks/'
for path in Path(path_of_the_directory).rglob('*.mp3'):
    f = os.path.join(path.parent,path.name)
    if os.path.isfile(f):
        ph = {}
        print("Filename: "+f)
        ph['filename'] = f
        audio = MP3(f, ID3=EasyID3)
        for tag in my_tags:
            if tag == "length":
                ph[tag] = str(audio.info.length)
                print(tag+": "+ph[tag])
            elif tag == "bitrate":
                ph[tag] = str(audio.info.bitrate/1000)+"K"
                print(tag+": "+ph[tag])
            elif tag == "comment":
                id3file = ID3(f)
                try:
                    ph[tag] = id3file['COMM::eng'].text[0]
                    print(tag+": "+str(id3file['COMM::eng'].text[0]))
                except KeyError:
                    ph[tag] = ""
                    pass
            elif tag == "filename":
                continue
            else:
                try:
                    ph[tag] = str(audio[tag][0])
                    print(tag+": "+str(audio[tag][0]))
                except KeyError:
                    ph[tag] = ""
                    pass
                except IndexError:
                    ph[tag] = ""
                    pass
        print("--------------------------------------------------")
        df.loc[len(df.index)] = [ph['title'], ph['artist'], ph['date'], ph['originaldate'], ph['album'], ph['tracknumber'], ph['discnumber'], ph['albumartist'], ph['compilation'], ph['genre'], ph['comment'], ph['length'], ph['bitrate']]
```

```python
sorted_df = df.sort_values(by='date')
sorted_df
```
