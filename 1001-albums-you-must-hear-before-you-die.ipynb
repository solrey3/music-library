```python
base_url = "http://www.1001albumsyoumusthearbeforeyoudie.net/albums-from-"

albums = []
artists = []
years = []
album_urls = []

for x in range(1955, 2019):
    url = base_url + str(x)
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    albums_cards = soup.find_all('table', class_='image-box')
    for album in albums_cards:
        artists.append(album.find('td',class_="image-box-heading").text)
        albums.append(album.find('td',class_="image-box-caption").text.rsplit(' (')[0])
        album_urls.append(album.find('a')['href'])
        years.append(str(x))
        print(album.find('td',class_="image-box-heading").text + " - " +\
            album.find('td',class_="image-box-caption").text.rsplit(' (')[0] + " (" +\
            str(x) + ") " +\
            album.find('a')['href'])
    
```

```python
url = "http://www.1001albumsyoumusthearbeforeyoudie.net/ex-albums-year"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
albums_cards = soup.find_all('table', class_='image-box')
for album in albums_cards:
    artists.append(album.find('td',class_="image-box-heading").text)
    albums.append(album.find('td',class_="image-box-caption").text.rsplit(' (')[0])
    album_urls.append(album.find('a')['href'])
    years.append(album.find('td',class_="image-box-caption").text.rsplit(' (')[1].replace(')',''))
    print(album.find('td',class_="image-box-heading").text + " - " +\
        album.find('td',class_="image-box-caption").text.rsplit(' (')[0] + " (" +\
        album.find('td',class_="image-box-caption").text.rsplit(' (')[1].replace(')','') + ") " +\
        album.find('a')['href'])
```

```python
wikipedia_titles = []

for album_url in album_urls:
    response = requests.get(album_url)
    soup = BeautifulSoup(response.text, 'html.parser')
    try:
        wikipedia_titles.append(soup.find('a',text="Wiki for mobiles view")['href'].rsplit('/')[-1])
        print(soup.find('a',text="Wiki for mobiles view")['href'].rsplit('/')[-1])
    except:
        wikipedia_titles.append("No wiki title available")
        print("No wiki title available")
    
```

```python
baseurl = "https://www.wikidata.org/w/api.php?"

release_dates = []

for title in wikipedia_titles:
    my_atts = {
        "titles": title,
        "action": "wbgetentities",
        "sites": "enwiki",
        "props": "claims",
        "format": "json",
    }
    try:
        resp = requests.get(baseurl, params = my_atts)
        data = resp.json()
        first_key = next(iter(data['entities']))
        release_dates.append(data['entities'][first_key]['claims']['P577'][0]['mainsnak']['datavalue']['value']['time'].split('T')[0].replace('+',''))
        print(title +" "+ data['entities'][first_key]['claims']['P577'][0]['mainsnak']['datavalue']['value']['time'].split('T')[0].replace('+',''))
    except:
        release_dates.append("N/A")
        print(title + " N/A")
```

```python
wiki_url = "https://en.wikipedia.org/wiki/New_Wave_(The_Auteurs_album)"
response = requests.get(wiki_url)
soup = BeautifulSoup(response.text, 'html.parser')
#print(soup.prettify())

release_date = soup.find('td', class_="published").text.rsplit('\xa0(')[0]
release_date
```

```python
baseurl = "https://en.wikipedia.org/wiki/"

release_dates = []

for title in wikipedia_titles:
    try:
        response = requests.get(baseurl + title)
        soup = BeautifulSoup(response.text, 'html.parser')
        release_dates.append(soup.find('td', class_="published").text.rsplit('\xa0(')[0].rsplit('[')[0])
        print(title +" "+ soup.find('td', class_="published").text.rsplit('\xa0(')[0].rsplit('[')[0])
    except:
        release_dates.append("N/A")
        print(title + " N/A")
```

```python
albums1001_df = pd.DataFrame({"Album": albums, "Artist": artists, "Year": years, "url_1001": album_urls, "wikipedia_title": wikipedia_titles, "release_date": release_dates})
albums1001_df.to_csv('albums1001.csv')
albums1001_df
```
