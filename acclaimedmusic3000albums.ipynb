```python
from bs4 import BeautifulSoup
import requests
import pymongo
from splinter import Browser
import pandas as pd
import time
import re
```

```python
url = "http://www.acclaimedmusic.net/year/alltime_albums.htm"
```

```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')

album_table = soup.find('table', class_='center')

album_table
```

```python
albums = album_table.find_all('tr')
```

```python
# print(albums[4].find_all('td')[1].text)
# print(albums[4].find_all('td')[2].text)
# print(albums[4].find_all('td')[4].text)
print(albums[4].find_all('td')[3].text.strip().replace("\r\n"," "))
```

```python
rank = []
artists = []
titles = []
years = []
genres = []

for x in range(1, 3001):
    rank.append(x)
    artists.append(albums[x].find_all('td')[1].text)
    titles.append(albums[x].find_all('td')[2].text)
    years.append(albums[x].find_all('td')[4].text)
    genres.append(albums[x].find_all('td')[3].text.strip().replace("\r\n"," "))
    print(str(x) +". "+ albums[x].find_all('td')[1].text +" - "+ albums[x].find_all('td')[2].text \
          +" ("+ albums[x].find_all('td')[4].text +") "\
          +" "+ albums[x].find_all('td')[3].text.strip().replace("\r\n"," "))
```

```python
acclaimed3000_df = pd.DataFrame({"Rank": rank, "Artist": artists, "Album": titles, "Year": years, "Genre(s)": genres})

acclaimed3000_df.set_index(['Rank'], inplace=True)

acclaimed3000_df.to_csv('acclaimedmusic3000albums.csv')

acclaimed3000_df
```
