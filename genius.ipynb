```python
import base64
import requests
import json
from config import genius_client_id
from config import genius_client_secret
from config import genius_client_access_token
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import linregress
import os
```

```python
base_url = 'https://api.genius.com'

search_path = 'search/'
search_request_uri = '/'.join([base_url, search_path])
#print(search_request_uri + user_input)

params = {'q': "chaka khan i feel for you"}

token = 'Bearer {}'.format(genius_client_access_token)
headers = {'Authorization': token}

r = requests.get(search_request_uri, params=params, headers=headers)
results = r.json()

#song_api_path 

song_result = results['response']['hits'][0]['result']['api_path']

song_result
```

```python
song_request_uri = base_url + song_result

s = requests.get(song_request_uri, params=params, headers=headers)
song_results = s.json()

print(song_results['response']['song']['title'])
print(song_results['response']['song']['primary_artist']['name'])
print(song_results['response']['song']['release_date'])
print(song_results['response']['song']['album']['name'])
```
