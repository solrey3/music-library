```python
from bs4 import BeautifulSoup
import requests
import pymongo
from splinter import Browser
import pandas as pd
import time
```

```python
url = "https://solrey3.github.io/music/2013/09/08/150ish-favorite-albums.html"
```

```python
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

album_table = soup.find('table')

album_table
```

```python
albums = album_table.find_all('tr')

albums
```

```python
print(albums[1].find_all('td')[0].text)
print(albums[1].find_all('td')[1].text)
print(albums[1].find_all('td')[1].a['href'])
print(albums[1].find_all('td')[2].text)
```

```python
artists = []
titles = []
wikipedia_urls = []
release_date = []

for x in range(1, 169):
    try:
        artists.append(albums[x].find_all('td')[0].text)
        titles.append(albums[x].find_all('td')[1].text)
        wikipedia_urls.append(albums[x].find_all('td')[1].a['href'])
        release_date.append(albums[x].find_all('td')[2].text)
    except:
        print("End of List: " + str(x))
```

```python
favoritealbums_df = pd.DataFrame({"artist": artists, "album": titles, "wikipedia_url": wikipedia_urls, "release_date": release_date})

favoritealbums_df.to_csv('favoritealbums_df.csv')
favoritealbums_df.to_json('favoritealbums_df.json', orient='records')

favoritealbums_df
```
