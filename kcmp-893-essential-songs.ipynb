```python
url = "https://www.thecurrent.org/list/893-essential-songs"

browser.visit(url)

rank = []
titles = []
artists = []
albums = []

for x in range(1, 10):
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    
    songs = soup.find_all('li', class_='countdown-item')
    
    for song in songs:
        rank.append(song.find('div', class_="countdown-rank").text.replace('.',''))
        titles.append(song.find('h3', class_="countdown-title").text.strip())
        artists.append(song.find('div', class_="countdown-artist").text.strip())
        albums.append(song.find('div', class_="countdown-album").text.strip())
        print(song.find('div', class_="countdown-rank").text.replace('.','') + ". " +\
              song.find('h3', class_="countdown-title").text.strip() + " - " +\
              song.find('div', class_="countdown-artist").text.strip() + " - " +\
              song.find('div', class_="countdown-album").text.strip())
    
    #browser.find_by_css('a.next').links.find_by_text('Next').click()
    try:
        browser.find_by_css('a.paginate_button.next').click()
    except:
        print("Last Page Reached")
    
```

```python
current893_df = pd.DataFrame({"Rank": rank, "Song": titles, "Artist": artists, "Albums": albums})
current893_df.set_index(['Rank'], inplace=True)
current893_df.to_csv('current893essential.csv')
current893_df
```
